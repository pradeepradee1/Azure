```python
%sql
DROP TABLE IF EXISTS orders_managed;

CREATE OR REPLACE TABLE orders_managed (
order_id BIGINT,
sku STRING,
product_name STRING,
product_category STRING,
qty INT,
unit_price DECIMAL(10,2)
);

ALTER TABLE orders_managed ADD CONSTRAINT valid_qty CHECK (qty > 0);
```
